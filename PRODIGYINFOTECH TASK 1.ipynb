```python
train_data=train_data[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath', 'SalePrice']]
test_data = test_data[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']]
# Feature engineering we will predict sales price based on glivarea-square footage,bedroom,fullbath and halfbath-bathroom
```

```python
train_data
```

```python
test_data
```

```python
# we will combine full bath and halfbath into total bath assuming 
train_data['TotalBath'] = train_data['FullBath'] + 0.5 * train_data['HalfBath']
test_data['TotalBath'] = test_data['FullBath'] + 0.5 * test_data['HalfBath']
```

```python
train_data
```

```python
test_data
```

```python
train_data = train_data.drop(['FullBath', 'HalfBath'], axis=1)
test_data = test_data.drop(['FullBath', 'HalfBath'], axis=1)
#since we have Total bath now we can remove fullbath and halfbath
```

```python
train_data
```

```python
# data is ready for machine learning   now we can start with feature and target in train_data
X=train_data[["GrLivArea",'BedroomAbvGr','TotalBath']]
Y=train_data["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split #train_test_split algorithm
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
le=LinearRegression()
```

```python
le.fit(X_train,Y_train)
```

```python
Y_predi=le.predict(X_train)
```

```python
Y_predi
```

```python
Y_predtest=le.predict(X_test)
```

```python
Y_predtest
```

```python
from sklearn.metrics import mean_squared_error,r2_score #evaluate metrics
```

```python
MSETRAIN=mean_squared_error(Y_train,Y_predi)
r2scoretrain=r2_score(Y_train,Y_predi)
print("mse of ALL on training:",MSETRAIN)
print("R2score of ALL on training :",r2scoretrain)
```

```python
MSETEST=mean_squared_error(Y_test,Y_predtest)
r2scoretest=r2_score(Y_test,Y_predtest)
print("mse of ALL on training:",MSETEST)
print("R2score of ALL on training :",r2scoretest)
```

```python
# well the accuracy says the model is showing moderate performance 
```

```python
from sklearn.linear_model import Ridge #L2
```

```python
L2=Ridge(1.0) #alphavalue in bracket
L2.fit(X_train,Y_train)
```

```python
L2.score(X_train,Y_train)
```

```python
L2.score(X_test,Y_test)
```

```python
L2.coef_
```

```python
L2.intercept_
```

```python
train_score = L2.score(X_train, Y_train)
test_score = L2.score(X_test, Y_test)
```

```python
for alpha in range(2,10):
    L2=Ridge(alpha)
    print("ridge_alpha:",alpha)
    print("training_score",train_score)
    print("testing_score",test_score)
```
